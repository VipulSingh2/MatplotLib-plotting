````python
import matplotlib.pyplot as plt
import numpy as np
x = ["Python","C","C++","Java"]
y = np.random.random((4))*100
````

````python
y
````

````python
plt.bar(x,y)
plt.show()
````

````python
plt.xlabel("language")
plt.ylabel("No.")
plt.title("Language vs votes")
plt.bar(x,y)
plt.show()
````

````python
plt.xlabel("language",fontsize = 10)
plt.ylabel("No.")
plt.title("Language vs votes")
plt.bar(x,y)
plt.show()
````

````python
plt.xlabel("language",fontsize = 15,color = 'green')
plt.ylabel("No.",fontsize = 15,color = 'green')
plt.title("Language vs votes",fontsize = 15,color = 'green')
plt.bar(x,y)
plt.show()
````

````python
z = [20,30,40,50]
p = np.arange(len(x))
p1 = [j + w for j in p]

w = 0.2
plt.xlabel("language",fontsize = 15,color = 'green')
plt.ylabel("No.",fontsize = 15,color = 'green')
plt.title("Language vs votes",fontsize = 15,color = 'green')
plt.bar(p,y,width = w,color='cyan', align = 'center', linewidth = 5,label = 'popularity')
plt.bar(p1,z,width = w,color='lime', align = 'center', linewidth = 5,label = 'popularity1')
plt.xticks(p+w/2,x)
plt.legend()
plt.show()
````
